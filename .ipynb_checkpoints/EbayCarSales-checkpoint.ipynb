```python
autos = pd.read_csv("autos.csv", encoding = 'Windows-1252')
autos.info()
autos.head()
```

```python
autos.columns
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odo_meter', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'num_of_pictures', 'postal_code',
       'last_seen']
```

```python
autos.head()
```

```python
autos.describe(include = 'all')
```

The num_of _pictures column needs more attention as it is showing strange values. Also the price and odo_meter columns contain non-numeric values.

```python
autos.info()
```

The following changes will be of removing non-character values from odo_meter and price.

```python
autos["price"] = autos["price"].str.replace("$","")
autos["price"]
```

```python
autos["odo_meter"] = (autos["odo_meter"]
                             .str.replace("km","")
                             .str.replace(",","")
                             .astype(int)
                             )
autos.rename({"odo_meter": "odometer_km"}, axis=1, inplace=True)
autos["odometer_km"].head()
```
